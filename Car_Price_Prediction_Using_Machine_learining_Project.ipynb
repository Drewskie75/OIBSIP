```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
Car_pred_dataset  =pd.read_csv('/content/car data.csv')
Car_pred_dataset.head()
```

```python
Car_pred_dataset.shape
```

```python
Car_pred_dataset.info()
```

```python
# checking the distribution of categorial data
print(Car_pred_dataset.Fuel_Type.value_counts())
print(Car_pred_dataset.Seller_Type.value_counts())
print(Car_pred_dataset.Transmission.value_counts())
print(Car_pred_dataset.Selling_Price.value_counts())
```

```python
#Encoding 'Fuel_Type' Column
Car_pred_dataset.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)

#Encoding 'Seller_Type' Column
Car_pred_dataset.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)

#Encoding 'Transmission_Type' Column
Car_pred_dataset.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
Car_pred_dataset.head()
```

Training Data

```python
X=Car_pred_dataset.drop(['Car_Name','Selling_Price'],axis=1)
Y=Car_pred_dataset['Selling_Price']
```

```python
X
```

```python
Y
```

```python
#Splitting the Training and Test data
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=1)
```

Model Training

```python
#loading the Linear regression model
lin_reg_Model=LinearRegression()
lin_reg_Model.fit(X_train,Y_train)
```

Model results

```python
#prediction on Training data
training_data_prediction=lin_reg_Model.predict(X_train)
```

```python
error_score=metrics.r2_score(Y_train,training_data_prediction)
print("R squared : ",error_score)
```

Visualize the actual prices and Predicted prices

```python
# @title
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Car prices vs Predicted Car prices")
plt.show()
```

```python
plt.style.use('_mpl-gallery')
fig = plt.figure(figsize = (10, 8))
plt.bar(Y_train,training_data_prediction,color='Blue')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Car prices vs Predicted Car prices")
plt.show()
```
